## 3.1. Преобразование вывода

В SQL реализованы операторы преобразования данных и встроенные функции, предназначенные для работы со значениями столбцов
и/или константами в выражениях. Использование этих операторов допустимо в запросах везде, где можно использовать выражения.

### Числовые, символьные и строковые константы

Несмотря на то, что SQL работает с данными в понятиях строк и столбцов таблиц, имеется возможность применения значений выражений, построенных с использованием встроенных функций, констант, имен столбцов, которые определяются как своего рода *виртуальные столбцы*. Они помещаются в списке столбцов и могут сопровождаться псевдонимами.

Если в запросе вместо спецификации столбца SQL обнаруживает число, то оно интерпретируется как *числовая константа*. Символьные константы указываются в одинарных кавычках. Если кавычка должна выводиться как часть строковой константы, то ее нужно предварить другой кавычкой.

<u>Пример</u>
```
SELECT 'Фамилия', SURNAME, 'Имя', NAME, 100, 'student''s office'
	FROM STUDENT limit 5;
```

### Арифметические операции для преобразования числовых данных

- Унарный (одиночный) оператор − (знак минус) изменяет знак числового значения, перед которым он стоит, на противоположный.
- Бинарные операторы +, −, * и / предоставляют возможность выполнения арифметических операций сложения, вычитания, умножения и деления.

<u>Пример</u>
```
select surname, name, stipend, kurs, -(stipend*kurs)/2 as calculation
	from student
	where kurs = 4 and stipend < 200;
```

### Символьная операция конкатенации строк

Операция конкатенации, обозначаемая символом “||”, позволяет соединять (“склеивать”) значения двух или более столбцов символьного типа или символьных констант в одну строку.

<u> Пример</u>
```
select surname || ', ' || name, stipend
	from student
	where kurs = 4 AND stipend > 200;
```

## 3.2. Скалярные функции

**IFNULL** — копия ненулевого аргумента.

```
ifnull(X,Y)
```

Функция ifnull(X,Y) возвращает копию ненулевого аргумента или NULL, если оба аргумента равны NULL.

<u> Пример</u><br>
У двух студентов - Сергеевой Фёклы и Тимофеевой Фаины - атрибут "Город" пуст, составим запрос, в котором заменим пустые значения на "Москва".

```
select student_id, surname, name, ifnull(city,'Москва')
    from student where city is null;
    
student_id  surname    name   ifnull(city,'Москва')
----------  ---------  -----  ---------------------
65          Сергеева   Фёкла  Москва
96          Тимофеева  Фаина  Москва
sqlite>
```

**LTRIM** — удаление левых граничных символов.
```
ltrim(<строка>[, <подстрока>])
```
- из <строки> удаляются слева символы, указанные в <подстроке>;
- если <подстрока> не указана, то по умолчанию удаляются пробелы;

**RTRIM** — удаление правых граничных символов.
```
rtrim(<строка>[, <подстрока>])
```

**SUBSTR** — выделение подстроки.

```
substr(<строка>, <начало>[, <количество>]
```
- из <строки> выбирается заданное <количество> символов, начиная с указанной позиции в строке <начало>;
- если <количество> не задано, символы выбирают с <начала> и до конца <строки>.

<u>Пример.</u>
```
SELECT substr(subj_name,1,6) FROM subject limit 5;

substr(subj_name,1,6)
---------------------
Анализ
Англий
Информ
Истори
Матема
```

**INSTR** — поиск подстроки в строке.
```
instr(<строка>, <подстрока>)
```
- тип возвращаемого значения — INT;
- функция возвращает позицию найденной подстроки.

<u>Пример.</u>

```
SELECT univ_name, instr(univ_name, 'У') FROM university LIMIT 5;

univ_name  instr(univ_name, 'У')
---------  ---------------------
ВГУ        3
НГУ        3
МФТИ       0
БГУ        3
ТГУ        3
sqlite>
```

- длина <строки>, тип возвращаемого значения — INT;
- функция возвращает NULL, если <строка> имеет NULL-значение.

## 3.3. Агрегатные функции

**AVG** — среднее значение.
```
avg(X) 
```

Возвращает среднее значение аргумента. Возвращаемое значение всегда вещественного типа.

<u>Пример</u>

```
sqlite> select avg(stipend) from student where kurs=5;
avg(stipend)
----------------
153.846153846154
sqlite>
```

**COUNT** — число значений.

count(X)
count(*)

Возвращает число отличных от нуля значений аргумента.

<u>Пример</u>

```
select count(*), count(stipend), count(city) from student;

count(*)  count(stipend)  count(city)
--------  --------------  -----------
89        89              87

```

**GROUP_CONCAT** — конкатенация группы значений.
```
GROUP_CONCAT(X,[ Y])
```
Возвращает строку, составленную из 'ненулевых' значений аргумента. Необязательный парметр Y определяет разделитель. По умолчанию используется разделитель - запятая (",").

<u> Пример</u>

```
SELECT univ_id,
       group_concat(quote(surname), ',') AS Surname,
       group_concat(stipend) AS Stipend
  FROM student
 WHERE univ_id IN (15, 11) 
 GROUP BY univ_id;

univ_id  Surname                                   Stipend
-------  ----------------------------------------  -----------
11       'Журавлева','Трофимов','Веселов','Жуков'  0,250,150,0
15       'Пак','Мельникова'                        100,250
```

**MAX** — максимальное значение.
```
max(X)
```
Возвращает максимальное значение в группе значений.

<u> Пример </u>
```
select max(stipend) from student;

max(stipend)
------------
250
```

**MIN** — минимальное значение.
```
min(X)
```
Возвращает минимальное значение в группе значений.

<u> Пример </u>
```
select min(kurs) from student;

min(kurs)
---------
1
```

**SUM** — сумма значений
```
sum(X)
```
Возвращает сумму значений.

<u> Пример </u>
```
select sum(stipend) from student;

sum(stipend)
------------
13210
```

Приведём пример на использовние нескольких агрегатных и скалярных функций в одном запросе.

<u>Пример</u>
```
SELECT min(rating), max(rating), round(avg(rating),2), count(univ_id)
	FROM university;

min(rating)  max(rating)  round(avg(rating),2)  count(univ_id)
-----------  -----------  --------------------  --------------
231          606          365.15                13
```

Агрегатная функция может быть использована совместно с ключевым словом 'DISTINCT'.

<u>Пример</u>
```
select count(distinct(city)) from university;

count(distinct(city))
---------------------
9
```

## 3.4. Неопределенные значения NULL в агрегатных функциях

### 3.4.1. Влияние NULL-значений в функции COUNT

Отсутствие значений в полях таблицы, обозначенное ключевым словом NULL, приводит к особенностям при выполнении агрегирующих
операций над данными, которые следует учитывать при их использовании в SQL-запросах.

Влияние NULL-значений в функции COUNT. Если аргумент функции COUNT является константой или столбцом без пустых значений, то функция возвращает количество строк, к которым применимо определенное условие или группирование.

Если аргументом функции является столбец, содержащий пустое значение, то COUNT вернет число строк, которые не содержат пустые
значения и к которым применимо определенное условие или группирование.

Если бы механизм NULL не был доступен, то неприменимые и отсутствующие значения пришлось бы исключать с помощью конструкции WHERE. Поведение функции COUNT(*) не зависит от пустых значений. Она возвратит общее количество строк в таблице.

### 3.4.2. Влияние NULL-значений в функции AVG.

Среднее значение множества чисел равно сумме чисел, деленной на число элементов множества. Однако если некоторые элементы пусты, т. е. их значения неизвестны или не существуют, то деление на количество всех элементов множества приведет к неправильному результату.

Функция AVG вычисляет среднее значение всех известных значений множества элементов, т. е. эта функция подсчитывает сумму
известных значений и делит ее на количество этих значений, а не на общее количество значений, среди которых могут быть NULL-значения. Если столбец состоит только из пустых значений, то функция AVG также возвратит NULL.

## 3.5. Группировка строк таблицы

### 3.5.1. Применение агрегатных функций в группе

Для объединения строк таблицы в группы используется предложение `GROUP BY` команды `SELECT`. Предложение `GROUP BY` позволяет применять агрегирующие функции к каждой группе, определяемой общим значением поля или полей, указанных в этом предложении.

```
SELECT exam_marks.subj_id, max(exam_marks.mark)
	FROM exam_marks
	GROUP by exam_marks.subj_id;

subj_id  max(exam_marks.mark)
-------  --------------------
10       5
11       1
12       5
13       5
22       5
43       3
56       5
73       4
94       5
```

В конструкции GROUP BY для группирования может быть использовано более одного столбца.
```
SELECT student_id, subj_id, max(mark)
    FROM exam_marks
    WHERE subj_id IN (11,12)
    GROUP BY student_id, subj_id;

student_id  subj_id  max(mark)
----------  -------  ---------
10          12       5
73          11       1
76          12       2
77          12       1
79          12       1
116         12       1
```

### 3.5.2. Фильтрация результатов группировки

Предложение `HAVING` команды `SELECT` позволяет отфильтровать результат группировки. Предложение `HAVING` определяет критерий, по которому группы следует включать в выходные данные (по аналогии с предложением `WHERE`, которое определяет критерий для отдельных строк).

<u>Пример</u>

```
SELECT semester, max(hour)
	FROM subject
	GROUP BY semester
	HAVING max(hour) between 34 and 70;

semester  max(hour)
--------  ---------
1         56
2         56
3         56
5         34
```

**JULIANDAY** — дата Юлианского календаря.
```
julianday(time-value, modifier, modifier, ...)
```
Возвращает десятичное число дней от 24 ноября 4714 года по полудни по Гринвичу до Р. Х.

История.

Древние египтяне измеряли год длительностью в 365 дней. Это меньше, чем астрономический год, поэтому начало года наступало раньше срока.

Римский император Юлий ввёл в употребление календарь, в котором каждый четвёртый год получал дополнительный, 366-й день. Таким образом, средний календарный год стал считаться равным 365,25 дня. Это уменьшило ошибку, но не полностью, ошибка получила противоположный знак. В результате начало года стало сдвигаться на более поздний срок, новый год наступал позже.

К концу XVI века сдвиг составил 10 дней и ошибка стала заметной. Римский папа ввёл календарь, в котором каждый четвертый год по-прежнему считался високосным, однако, каждый 100-й год не считался високосным, если он не делился нацело на 400. Таким образом, 2000-й год високосный, а 1900-й и 2100-й года — нет. Так появился Григорианский календарь, в соответствии с которым в году в среднем 365,2422 дня.

Более подробно:
https://pumas.nasa.gov/sites/default/files/examples/04_21_97_1.pdf

<u>Пример</u>

```
select julianday('2022-03-10 15:00:00');

julianday('2022-03-10 15:00:00')
--------------------------------
2459649.125
```

<u>Пример</u>

Найдём, сколько дней прошло со дня Великой Отябрьской Социалистической Революции

```
select julianday('now') - julianday('1917-11-07');

julianday('now') - julianday('1917-11-07')
------------------------------------------
38108.6368640163
```

## 3.8. Упорядочение выходных полей 

Как уже отмечалось, записи в таблицах реляционной базы данных неупорядочены. Однако, данные, выводимые в результате выполнения запроса, могут быть упорядочены. Для этого в операторе SELECT используется предложение с ключевым словом ORDER BY, которое позволяет упорядочивать выводимые записи в соответствии со значениями одного или нескольких выбранных столбцов. При этом можно задать возрастающую (ASC) или убывающую (DESC) последовательность сортировки для каждого из столбцов. По умолчанию принята возрастающая последовательность сортировки. Запрос, позволяющий выбрать все данные из таблицы предметов обучения SUBJECT, с упорядочением по наименованиям предметов, выглядит следующим образом:
```
SELECT *
    FROM SUBJECT
    ORDER BY SUBJ_NAME;
```
Тот же список, но упорядоченный в обратном порядке, можно
получить запросом:
```
SELECT *
    FROM SUBJECT
    ORDER BY SUBJ_NAME DESC;
```
Можно упорядочить выводимый список предметов обучения по значениям семестров, а внутри семестров — по наименованиям предметов:
```
SELECT *
    FROM SUBJECT
    ORDER BY SEMESTR, SUBJ_NAME;
```
Предложение ORDER BY может использоваться с GROUP BY. При этом оператор ORDER BY указывается последним:
```
SELECT SUBJ_NAME, SEMESTR, MAX(HOUR)
    FROM SUBJECT
    GROUP BY SEMESTR, SUBJ_NAME
    ORDER BY SEMESTR;
```
При упорядочивании вместо наименований столбцов можно указывать их номера, имея, однако, в виду, что в данном случае это номера
столбцов, указанные при определении выходных данных в запросе, а не номера столбцов в таблице. В запросе
```
SELECT SUBJ_ID, SEMESTR
    FROM SUBJECT
    ORDER BY 2 DESC;
```
выводимые записи будут упорядочены по полю SEMESTR.

Если в поле, которое используется для упорядочивания, существуют NULL-значения, то они размещаются в конце или предшествуют всем остальным значениям этого поля.
