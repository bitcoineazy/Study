```python
#Цикл
def factorial(n):
    fact = 1
    for num in range(2, n + 1):
        fact *= num
    print(fact)
```

```python
#Рекурсия
def factorial(n):
    if n < 2:
        return 1
    else:
        return n * factorial(n-1)
```

```python
def factorial_r(n):
    global x
    fact = 1
    k = 0
    x = n
    sp_amount = 0

    for num in range(2, n + 3):
        fact *= num
        k += 1
        print(f'{"    " *k}factorial({x})')
        x -= 1
        sp_amount += 4

    fact = 1
    print(f'{" "*sp_amount}{fact}')
    for reversed in range(1, n + 1):
        fact *= reversed
        print(f'{" "*(sp_amount-k)}{fact}')
        x += 1
        k += 3
```

```python
factorial_r(5)
```
